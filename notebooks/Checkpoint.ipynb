# Continuous Training using ensemble networks


Artificial Neural Networks classifying together to ensure robustness by correcting each other. 

## What is the research question at hand?

 - How accurate do artificial neural networks have to be, to efficiently train each other in an image recognition task?

## Theoretical Approach

What assumptions can be made about the efficiency of the method by looking at the probabilities?

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(4, 1, figsize=(15,42))
x = np.linspace(0,1)
postrain3 = 3 * x**2 * (1-x)
negtrain3 = 3 * (1-x)**2 * x
postrain5 = 5 * x**4 * (1-x)
negtrain3 = 3 * (1-x)**2 * x
negtrain5 = 5 * (1-x)**4 * x
miss5 = 10* ((1-x)**3*x**2 + x**3*(1-x)**2)
allhit5 = x**5
allmiss5 = (1-x)**5
plt.suptitle("Theoretical Analysis of the method")
ax[0].set_title("Probability distribution of ensemble training")
ax[1].set_title("Probability distribution of ensemble training")
ax[2].set_xlabel("Accuracy of the networks")
ax[0].set_xlabel("Accuracy of the networks")
ax[2].set_ylabel("Likelihood")
ax[0].set_ylabel("Likelihood")
ax[3].set_xlabel("Accuracy")
ax[3].set_ylabel("Likelihood")
ax[0].plot(x, postrain3)
ax[1].plot(x, postrain5)
ax[0].plot(x, negtrain3)
ax[1].plot(x, negtrain5)
ax[2].set_title("Probability distributions of all possible outcomes")
ax[2].plot(x, miss5)
ax[2].plot(x, allhit5)
ax[2].plot(x, allmiss5)
ax[2].plot(x, postrain5)
ax[2].plot(x, negtrain5)
ax[2].plot(x, miss5+allhit5+allmiss5+negtrain5+postrain5)
ax[3].plot(x, allhit5+postrain5)
ax[3].plot(x, allmiss5+negtrain5)
ax[3].plot(x, miss5)
ax[3].set_title("Predictions plus training")

#plot dotted line at 0.7
ax[0].axvline(0.7, linestyle='--')
ax[1].axvline(0.7, linestyle='--')

"""
# Plot the red markes
for point in [0.3, 0.7]:
    value3 = 3*point**2*(1-point)
    value5 = 5*point**4*(1-point)
    ax[0].plot(0.7, value3, 'ro')
    ax[0].plot(0.7, value5, 'bo')
    ax[0].annotate(str(value3),xy=(0.7,value3))
    ax[0].annotate(str(value5),xy=(0.7,value5))
"""
    
ax[0].grid(True)
ax[1].grid(True)
ax[2].grid(True)
ax[3].grid(True)
ax[0].legend(['3 Networks Positive Training', '3 Networks Negative Training', 'x = 70%'])
ax[1].legend(['5 Networks Positive Training', '5 Networks Negative Training', 'x = 70%'])
ax[2].legend(['No Prediction', 'All right', 'All wrong', 'Positive Training', 'Negative Training', 'Sum'])
ax[3].legend(['Correct Prediction', 'Wrong Prediction', 'No Prediction'])
plt.show()
```
